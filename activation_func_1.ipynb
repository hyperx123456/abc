### Write a Python program to plot a few activation functions that are being used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,50)
x
```

## 1. ReLU function

```python
plt.plot(x, np.maximum(0,x),label='Relu')
```

## 2. Tanh function

```python
plt.plot(x, np.tanh(x), label='tanh')
```

## 3. Softmax function

```python
plt.plot(x, np.exp(x)/np.sum(np.exp(x)), label='softmax')
```

## 4. Linear function

```python
plt.plot(x, 5*x, label='linear')
```

## 5. Swiss Function

```python
plt.plot(x, x*(1/(1+ np.exp(-x))), label='swiss')
```

## 6. Sigmoid funtion

```python
plt.plot(x, 1/(1+ np.exp(-x)), label='swiss')
```

## 7. Bipolar Sigmoid Function

```python
plt.plot(x, (1- np.exp(-x))/(1+ np.exp(-x)), label='swiss')
```
